```python
import keras
```

```python
inputs = keras.Input(shape=(100,))
print(inputs)
x1 = keras.layers.Dense(10, activation='relu')(inputs[:,:50])
x2 = keras.layers.Dense(10, activation='relu')(inputs[:,50:])
model   = keras.Model(inputs=inputs, outputs=[x1,x2])
model.summary()
```

```python
keras.utils.plot_model(model, to_file='test.png', show_shapes=True)
```
